```python
import pandas as pd
import matplotlib.pyplot as plt

def dividir_ventana(df, T, H):
    for i in range(len(df) - T - H + 1):
        train_start_index = i
        train_end_index = i + T - 1
        test_index = i + T

        df.loc[train_start_index:train_end_index, 'set'] = 'train'
        df.loc[test_index:test_index + H - 1, 'set'] = 'test'
        df.loc[train_start_index - 1:train_start_index, 'set'] = 'NaN'

        # Visualizacin grfica
        plt.figure(figsize=(10, 6))
        plt.plot(df['fecha'], df['producto1'], label='Producto 1')
        plt.plot(df['fecha'], df['producto2'], label='Producto 2')
        plt.scatter(df.loc[df['set'] == 'train', 'fecha'], df.loc[df['set'] == 'train', 'producto1'], color='red', label='Train')
        plt.scatter(df.loc[df['set'] == 'test', 'fecha'], df.loc[df['set'] == 'test', 'producto1'], color='green', label='Test')
        plt.title(f'Iteracin {i + 1} - Window: {df.loc[train_start_index:test_index + H - 1, "fecha"].tolist()}')
        plt.legend()
        plt.show()

    # Imprimir los resultados finales
    print("\nResultados finales:")
    print(df)

# Datos proporcionados
data = {
    'fecha': ['2008-01-31', '2008-02-29', '2008-03-31', '2008-04-30', '2008-05-31', '2008-06-30', '2008-07-31', '2008-08-31', '2008-09-30', '2008-10-31'],
    'producto1': [4000.00, 3997.40, 3978.61, 3986.45, 3979.70, 3974.22, 3984.10, 3981.70, 3976.71, 3968.33],
    'producto2': [3000.00, 3010.69, 3022.02, 3033.92, 3038.40, 3047.62, 3049.90, 3061.83, 3077.36, 3086.42]
}

df = pd.DataFrame(data)

# Definir T (tamao de la ventana) y H (tamao fuera de la ventana)
T = 3
H = 2

# Llamar a la funcin
dividir_ventana(df, T, H)
```
